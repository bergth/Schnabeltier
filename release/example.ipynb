## Environnement

Definir un environnement:
```python
env (% temp, % hum, % car, % eau)
```

````python
e = env(50,50,50,50)
````

## Individu
Définir un individu (facultatif):
```python
ind (% pilosité, % palmation, % herbivore, % carnivore)
```

````python
i = ind(50,50,50,50)
````

## Lancer la simulation
Si l'individu n'est pas utilisé, les individus seront définis aléatoirement
Sinon ils seront tous similaires
```python
vogel(environnement, Individu = None)
```

````python
V = vogel(e,i)
````
